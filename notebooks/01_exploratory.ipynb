# Exploratory analysis

Use this notebook for one-off looks at the data. **Do not put pipeline logic here** — that lives in `scripts/`. Notebooks are for sanity-checking and quick visuals, not for generating final manuscript numbers.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
import pandas as pd
from src.config import SCORES_DIR

originals = pd.read_csv(SCORES_DIR / 'originals.csv')
originals.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

if 'site' in originals.columns and len(originals) > 1:
    fig, ax = plt.subplots(figsize=(8, 5))
    sns.boxplot(data=originals, x='site', y='fkgl', ax=ax)
    ax.axhline(6.0, color='red', linestyle='--')
    plt.xticks(rotation=30, ha='right')
    plt.show()
```
